```python
import sys
import os
import glob
sys.path.append("..")  # add project root (parent of examples) to Python path

DATA_DIR = "../data"

from mizan_vector import MizanMemoryStore, HFEmbedder

def load_text_files(data_dir: str):
    docs = []
    paths = []
    for path in glob.glob(os.path.join(data_dir, "*.txt")):
        with open(path, "r", encoding="utf-8") as f:
            text = f.read().strip()
        if not text:
            continue
        docs.append(text)
        paths.append(os.path.basename(path))
    return paths, docs

file_names, docs = load_text_files(DATA_DIR)
print(f"Loaded {len(docs)} documents")

#intfloat/e5-base-v2
#sentence-transformers/paraphrase-MiniLM-L12-v2
embedder = HFEmbedder("intfloat/e5-base-v2")  # or any other model

# 2) Create a memory store
store = MizanMemoryStore(dim=384)  # 384 = dimension of that HF model

# 1) Embed all documents
embs = embedder.encode(docs)   # same HFEmbedder you already created

# 2) Add them to the store
store = MizanMemoryStore(dim=len(embs[0]))

for fname, content, emb in zip(file_names, docs, embs):
    # adapt if your add_document supports extra metadata, e.g. metadata={"file": fname}
    store.add_document(content=content, embedding=emb)
```

```python
query = "who is ahsan"
q_emb = embedder.encode_one(query)

results = store.search(q_emb, top_k=3, metric="mizan")

print(f"Query: {query}")

for r in results:
    print(f"Score: {r.score:.3f} | Text: {r.content}")
```

```python
query = "ahsan?"
q_emb = embedder.encode_one(query)

print("=== Question: ===")
print(f"{query}")

print("=== MIZAN similarity ===")
mizan_results = store.search(q_emb, top_k=5, metric="mizan")
for r in mizan_results:
    print(f"[mizan] {r.score:.3f} | {r.content[:250]}")

print("\n=== COSINE similarity ===")
cos_results = store.search(q_emb, top_k=5, metric="cosine")
for r in cos_results:
    print(f"[cos]   {r.score:.3f} | {r.content[:250]}")
```
